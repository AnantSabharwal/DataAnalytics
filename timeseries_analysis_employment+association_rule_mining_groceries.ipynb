```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sb
%matplotlib inline
```

```python
df = pd.read_csv("employment.csv")
df
```

# Q4

```python
ax = plt.figure(figsize = (30,20))
plt.xlabel("Months",fontsize = 12)
plt.ylabel("Unemployment rate",fontsize = 12)
for x in employment.columns:
    plt.plot(employment[x],linewidth = 5,label=employment.columns)
#add x and y labels
# add legends
```

```python
employment.plot(subplots=True, figsize=(15, 15))
```

# Q5

```python
#yearly trends
df_year=employment.resample("Y").mean()
df_year.plot( figsize=(15,15))
```

```python
#monthly trends 
df_month=employment.resample("M").mean()
df_month.plot( figsize=(15,15))
```

```python
employment["total_emp"] = employment.sum(axis=1)/(len(list(employment))-1)
emp_month = employment.resample("M").mean()
fig, ax = plt.subplots(figsize=(15,6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.bar(emp_month['2000':].index, emp_month.loc['2000':, "total_emp"], width=25, align='center')
plt.xlabel("Months",fontsize = 12)
plt.ylabel("Unemployment rate",fontsize = 12)
plt.grid()
plt.show()
```
