# MetaData
- name = cereal name
- mfr = manufacturer
- type = hot/cold
- calories = number of calories
- protein = proteing(g)
- fat = sodium (g)
- sodium = sodium (mg)
- fiber = dietary fiber(g)
- carbo = complex carbohydrates(g)
- sugars = sugar(g)
- shelf = display shelf (1, 2, or 3, counting from the floor)
- potass = potassium(mg)
- vitamins = 0, 25, or 100, respectively indicating 'none added'; 'enriched, often to 25% FDA recommended'; '100% of FDA --recommended
- weight = serving size
- cups = cups per serving
- A=American Home Food Products, G=General Mills, K=Kelloggs, N=Nabisco, P=Post, Q=Quaker Oats, R=Ralston Purina

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

# reading data sets
cereals = pd.read_excel("Cereals1.xls")
cereals
```

# Q1

```python
# table of the 5 number summary of all the numeric attributes
```

```python
summary = {'Attribute':list(cereals.columns.values)[3:16],'Min':[],'Max':[],'Median':[],'2nd Quarile':[],'3rd Quartile':[],'Mean':[]}
for i in range(3,16):
    summary['Min'].append(cereals.iloc[:,i].min())
    summary['Max'].append(cereals.iloc[:,i].max())
    summary['Median'].append(cereals.iloc[:,i].median())
    summary['2nd Quarile'].append(cereals.iloc[:,i].quantile(q=0.75))
    summary['3rd Quartile'].append(cereals.iloc[:,i].quantile(q=0.25))
    summary['Mean'].append(cereals.iloc[:,i].mean())
cereals_tem=pd.DataFrame(summary)
cereals_tem
```

# Q4

```python
# filtering noisy data and replacing by median
NutrientList = ['protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']
for i in NutrientList:
    Q1 = cereals[i].quantile(0.25)
    Q3 = cereals[i].quantile(0.75)
    IQR = Q3 - Q1
    O1 = Q1 - (1.5*IQR)
    O2 = Q3 + (1.5*IQR)
    cereals.loc[(cereals[i] < O1) | (cereals[i] > O2), i] = cereals[i].median()
cereals
```

# Q13

```python
# relationship between manufacturer and rating
#Convert rating to categorical variable
cereals.rating.fillna(cereals.rating.mean(),inplace = True)
cereals['OG_Rating'] = cereals.rating 
cereals.rating = (cereals.rating/10).astype('int')
#Convert Ratings column from a float column to 10-point form
```

```python
# We use the chi2_contingency function from scipy.stats. Scipy is a Python library used for scientific computing
# and technical computing. SciPy contains modules for optimization, linear algebra, integration, interpolation, 
# and other tasks common in science and engineering.
from scipy.stats import chi2_contingency
```

```python
# We repeat chi-square test since we now have to find whether there's a relationship between mfr and ratings 
#both of which are categorical variables
chi_square_test = pd.crosstab(index = cereals.rating,columns = cereals.mfr)
c, p, dof, expected = chi2_contingency(chi_square_test) 
p
```

```python
# there the p value is much smaller than the 0.05 threshold 
# P â‰¤ 0.05 means that the test hypothesis is false or should be rejected
```
