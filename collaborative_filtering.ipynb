```python
# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
ratings_df = pd.read_csv("ratings.csv",index_col = "userId")
del ratings_df['timestamp']
ratings_df
```

```python
#creating a pivot table
ratings_pivot = ratings_df.pivot_table(values = 'rating',index = 'userId',columns = 'movieId')
ratings_pivot
```

```python
ratings_pivot.fillna(0,inplace=True)
ratings_pivot
```

```python
#importing sklearn libraries that are needed
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
from scipy.spatial.distance import correlation
```

```python
A = np.array(ratings_pivot)
dist_out =1- pairwise_distances(A,metric = "cosine")
dist_out
```

```python
#creating the above generated array into a dataframe
sim = pd.DataFrame(dist_out)
sim
```
